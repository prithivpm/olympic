```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset
data = pd.read_csv('athlete_events.csv')

# Display basic information about the dataset
print("Dataset Information:")
print(data.info())

# Display summary statistics
print("\nSummary Statistics:")
print(data.describe())

# Display the first few rows of the dataset
print("\nFirst Few Rows:")
print(data.head())

# Visualize the distribution of medals over the years
plt.figure(figsize=(12, 6))
sns.countplot(x='Year', data=data, hue='Medal', palette='viridis')
plt.title('Distribution of Medals Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()

# Visualize the distribution of medals by sport
plt.figure(figsize=(16, 8))
sns.countplot(x='Sport', data=data, hue='Medal', palette='viridis')
plt.title('Distribution of Medals by Sport')
plt.xlabel('Sport')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()

# Visualize the distribution of medals by country
plt.figure(figsize=(20, 10))
sns.countplot(x='NOC', data=data, hue='Medal', palette='viridis', order=data['NOC'].value_counts().index)
plt.title('Distribution of Medals by Country')
plt.xlabel('Country Code (NOC)')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

# Display the top countries with the most medals
top_countries = data[data['Medal'].notnull()]['NOC'].value_counts().head(10)
print("\nTop Countries with the Most Medals:")
print(top_countries)

# Visualize the age distribution of athletes
plt.figure(figsize=(12, 6))
sns.histplot(data['Age'].dropna(), bins=20, kde=True, color='purple')
plt.title('Age Distribution of Athletes')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
